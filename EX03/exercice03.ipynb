## Step 1 : Computing the upper bound

In this exercise, the objective is to determine the upper bound of the complexity for the provided functions, specifically function_1() and function_2(). The assignment also emphasizes that it is preferable to have an O(n^2) bound rather than an O(n^4) bound. In other words, the goal is to achieve the smallest possible complexity.

Function1:

```python
def function_1(n: int) -> None:
    intermediate_list = []
    total = 0
    for i in range(n * 2):
        total += i
        intermediate_list.append(total)
    sum(intermediate_list)

def compute_upper_bound(n):
    return n**2

# Impl:
n = 10
function_1(n)
print(compute_upper_bound(n))
```

Function2:

```python
def function_2(n: int) -> None:
    import random
    temp_list = [random.randint(0, n) for _ in range(n)]
    temp_list.sort()
    max(temp_list)

def compute_upper_bound(n):
    return n**2

# Impl:
n = 10
function_2(n)
print(compute_upper_bound(n))
```

## Step 2 : Verify the bound

In this second part, we need to verify if the bound is correct by fitting a polynomial of the found degree to a
curve representing the measured elapsed computation time as a function of n.
For instance in figure 3, the function as a O(n3 ) complexity, so a fitted polynomial of degree 2 can not represent well the curve.

```python
from random import shuffle
import numpy as np
import timeit
import matplotlib.pyplot as plt

def analyze_complexity_1(size: int) -> None:
    
    accumulation = []
    for index in range(size**2):
        subtotal = 0
        for subindex in range(index):
            subtotal += subindex
        accumulation.append(subtotal)
    total_sum = sum(accumulation)
    return total_sum

# Estimate the upper bound for the computational complexity
def estimate_complexity_limit(size):
    return size**4

# Measure the time taken for different input sizes
input_sizes = np.arange(1, 10, 1)  # Adjust the range as required
time_records = []
for size in input_sizes:
    time_taken = timeit.timeit(lambda: analyze_complexity_1(size), number=1)
    time_records.append(time_taken)

# Fit a polynomial of degree 4 to the recorded times
poly_coefficients = np.polyfit(input_sizes, time_records, 3)

# Plot the recorded times against input sizes
plt.plot(input_sizes, time_records, 'o', label='Recorded times')
plt.xlabel('Input size')
plt.ylabel('Time taken')
plt.title('Time complexity as a function of input size')
plt.grid(True)

# Generate and plot polynomial values
fitted_curve = np.polyval(poly_coefficients, input_sizes)
plt.plot(input_sizes, fitted_curve, '-', label=f'Fitted polynomial: {poly_coefficients}')
plt.legend()
plt.show()
```

The bound i is correct
